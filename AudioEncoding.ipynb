```python
%pip install pyttsx3
import pyttsx3

def text_to_wav(text, output_path="output.wav"):
    engine = pyttsx3.init()
    
    # optional voice/volume/rate adjustments
    engine.setProperty('rate', 170)    # speed
    engine.setProperty('volume', 0.9)  # loudness

    engine.save_to_file(text, output_path)
    engine.runAndWait()
    print(f"WAV file generated: {output_path}")

# Example usage:
text = "Now we shall discuss Lab Number 1 — the Advanced Software and Distributed Computing Lab, where cutting-edge software engineering, big data analytics, and distributed systems innovation take place."
text_to_wav(text, "speech.wav")
```
